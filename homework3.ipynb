```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Dice:
    def flip(self):
        return random.randint(1, 6)
```

```python
dice = Dice()
```

```python
def my_expacted_value(x):
    p = len(x) ** - 1
    summ = 0
    for i in x:
        summ += i * p
    return summ
```

```python
def my_mean(x):
    return sum(x)/len(x)
```

```python
def my_variance(x):
    return sum([(xi - my_mean(x))**2 for xi in x]) / (len(x) - 1)
```

```python
dice_numbers = [i for i in range(1, 7)]
print(my_expacted_value(dice_numbers))
print(my_variance(dice_numbers))
```

```python
dice_flips = [dice.flip() for i in range(1000000)]
print(my_expacted_value(dice_flips))
print(np.var(dice_flips))
```

```python
def flip_n(n):
    summ = 0
    for i in range(n):
        summ += dice.flip()
    return summ
```

```python
f2 = [flip_n(2) for i in range(200000)]
f4 = [flip_n(4) for i in range(200000)]
f6 = [flip_n(6) for i in range(200000)]
f8 = [flip_n(8) for i in range(200000)]
f10 = [flip_n(10) for i in range(200000)]
f = f2 + f4 + f6 + f8 + f10
plt.hist(f)
plt.show()
```
